```
#Complete major project

import random
player_score={"Kirat Boli":[0,0],"N.S. Nodhi":[0,0],"R Rumrah":[0,0],"Shashi Henra":[0,0]}
player_list=["Kirat Boli","N.S. Nodhi","R Rumrah","Shashi Henra"]
run_list=[0,1,2,3,4,5,6,-1]
player_stats={"Kirat Boli":[5,30,25,10,15,1,9,5],"N.S. Nodhi":[10,40,20,5,10,1,4,10],"R Rumrah":[20,30,15,5,5,1,4,20],"Shashi Henra":[30,25,5,0,5,1,4,30]}
x=1 
strike="Kirat Boli"
non_strike="N.S. Nodhi"
for i in range(0,4):
  for j in range(0,6):
    for l in player_list:
       if l==strike:
             random_variable=random.choices(run_list,player_stats[strike],k=1)
             
             if random_variable[0]==-1:
                x=x+1
                player_score[strike][1]=player_score[strike][1]+1
                print("out",strike,"at",i,".",j+1)

                if x<4:
                     strike=player_list[x]
                     print("new player opt to bat",strike)
                else:
                   strike=""
                   break
                
             else:
                if random_variable[0]%2==0:
                   player_score[strike][0]=player_score[strike][0]+random_variable[0]
                   player_score[strike][1]=player_score[strike][1]+1
                   print(strike,"hits",random_variable,"run","at",i,".",j+1)
                else:
                   player_score[strike][0]=player_score[strike][0]+random_variable[0]
                   player_score[strike][1]=player_score[strike][1]+1
                   print(strike,"hits",random_variable,"run","at",i,".",j+1)
                   temp=strike
                   strike=non_strike
                   non_strike=temp
             break                 
    if x>3:
         break                 
  print("-------------",i+1,"over finish---------------")    
  if x<3:
    if i<3:
      temp=strike
      strike=non_strike
      non_strike=temp
    else:
       print("--------------------------------End Match----------------------------------")
  else:
      print("--------------------------------End Match----------------------------------")
      print("--------------All out------------------")
      break

  
    

  
   

for m,n in player_score.items():
    if m==strike:
       print(m,":",n[0],"* (",n[1],")") 
    else:   
       print(m,":",n[0],"(",n[1],")")  
```
